%matplotlib inline will create static images of your plots embedded in the notebook.

## Visualization  Techniques

### Scatter Plots

```python
x = np.random.randn(100)
y = np.random.randn(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)
```

```python
plt.scatter(x,y,c=colors,s=sizes,alpha=0.2,cmap='viridis')
plt.colorbar();
```

### Line Plots

```python

year = [1970 , 1980, 1990, 2000, 2010, 2020]
pop_A = [4.9, 5.3, 7.1, 10.7, 13.5, 15.6]
pop_B = [44.4, 55.6, 69.7 , 87.1 , 95.4 , 100.5]
```

```python
plt.plot(year, pop_A, color='g')
plt.plot(year, pop_B, color='r')
plt.xlabel('Countries')
plt.ylabel('Popiltion in millions')
plt.title('Population over the years')
```

### Historgrams

```python
data = np.random.randn(1000)
plt.hist(x=data)
```

```python
plt.hist(data, bins=30, normed=True, alpha=0.5,
                histtype='stepfilled', color='steelblue',
                edgecolor='none');
```

### Bar Plots

```python
labels = ('Python', 'C++' , 'Java' , 'Pearl', 'C#')
num_users = [12,8,15,4,6]
index = np.arange(len(labels))
plt.bar(index,num_users,align='center',alpha=0.5)
plt.xticks(index,labels)
plt.xlabel('Language')
plt.ylabel('Num of Users')
plt.title('Programming Language Usage')
```

### Plot Boxes

```python
data_group1 = np.random.normal(100, 10, 200)
data_group2 = np.random.normal(80, 30, 200)
data_group3 = np.random.normal(90, 20, 200)

# combine these different datasets into a list    
data_to_plot = [data_group1, data_group2, data_group3]

# Create the boxplot
# patch_artist must be True to control box fill (color-filled boxes)
plt.boxplot(data_to_plot, patch_artist = True)

plt.xlabel('Group')
plt.ylabel('Scores')
plt.title('Some Group Scores')
```
